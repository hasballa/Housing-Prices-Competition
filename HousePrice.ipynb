```python
# Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
df = pd.read_csv('sample_submission.csv')
```

```python
train_set.columns[train_set.isna().any()]
d = ['LotFrontage', 'MasVnrType', 'MasVnrArea', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Electrical', 'GarageType', 'GarageYrBlt',
       'GarageFinish', 'GarageQual', 'GarageCond']
train_set[d]= train_set[d].fillna(train_set[d].mean())
train_set[d]
```

```python
train_set.drop(['Alley', 'FireplaceQu', 'PoolQC','Fence', 'MiscFeature'], axis = 1,inplace = True)
```

```python
train_set.head(5)
```

```python
train_set.info()
```

```python
t = train_set.select_dtypes(include ='object').columns
train_set[t] = train_set[t].fillna('Others')
```

```python
#Converting Categorical Columns
from sklearn.preprocessing import OrdinalEncoder
OrdinalEncoder = OrdinalEncoder()
train_set[t] = OrdinalEncoder.fit_transform(train_set[t])
```

```python
train_set[t]
```

```python
train_set.head()
```

```python
#Normalizing Numerical Columns
from sklearn.preprocessing import StandardScaler

standardscaler = StandardScaler()
train_set.iloc[:, 0:75] = standardscaler.fit_transform(train_set.iloc[:, 0:75])
train_set.iloc[:, 0:75]
```

```python
# Splitting Data

X = train_set.iloc[:, 0:75]
y = train_set.iloc[:, -1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.05, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 277, random_state = 0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```
